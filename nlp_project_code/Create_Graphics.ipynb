```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pandas as pd
sns.set(style='ticks', palette='Set2')
import os
import numpy as np
%matplotlib inline
```

```python
def to_np(x):
    return np.array([float(v.replace("[","").replace("]","").replace(",","")) for v in x.split()])
def get_data(dir_in = "../data/predictions/"):
    mean_list = []
    std_list = []
    for f in os.listdir(dir_in):
        if '.csv' in f:
            word = f.split(".")[0]
            type_word = word.split("-")[1]
            df = pd.read_csv(dir_in + f, index_col=0)
            df = df.applymap(to_np)
            df_mean = df.applymap(np.mean)
            df_std = df.applymap(np.std)
            df_mean['word'] = word
            df_mean['word_type'] = type_word
            mean_list.append(df_mean)
            df_std['word'] = word
            df_std['word_type'] = type_word
            std_list.append(df_std)
    all_mean = pd.concat(mean_list)
    all_std = pd.concat(std_list)
    return all_mean, all_std
def make_plot(word, dir_out = "../graphics/plots/", res = 1000, mean_df=mean_df,
             std_df=std_df, y_min=0, y_max=1):

    order = ['syntax', 'semantics', 'combined']
    col_order = ['randomForest', 'svm', 'both_mean', 'both_max']

    m = mean_df[mean_df['word'] == word]
    m = m[col_order]
    s = std_df[std_df['word'] == word]
    s = s[col_order]
    n = 4
    ind = np.arange(n)
    width = 0.25

    fig, ax = plt.subplots(figsize=(8,5))
    rects1 = ax.bar(ind, m.ix[order[0], :4], width, color='r')
    rects2 = ax.bar(ind+width, m.ix[order[1], :4], width, color='g')
    rects3 = ax.bar(ind+(width*2), m.ix[order[2], :4], width, color='b')
    #ax.set_title(word)
    ax.set_ylabel('Average Accuracy')
    ax.set_ylim(y_min, y_max)
    ax.set_xticks(ind+width)
    ax.set_xticklabels( ('Random Forest', 'SVM', 'Ensemble 1', 'Ensemble 2') )
    # Shrink current axis by 20%
    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
    # Put a legend to the right of the current axis
    ax.legend( (rects1[0], rects2[0], rects3[0]), ('Syntactic Features', 'Semantic Features', 'All Features'),
             loc='center left', bbox_to_anchor=(1, 0.5))
    sns.despine()
    plt.savefig(dir_out + '{}.png'.format(word), dpi=res)
def t_test(word, model_ind):
    t_df = pd.read_csv("../data/predictions/{}.csv".format(word), index_col=0)
    t_df = t_df.applymap(to_np)
    v_1 = t_df.ix[0,model_ind]
    v_2 = t_df.ix[2,model_ind]
    return stats.ttest_ind(v_1, v_2)
```
